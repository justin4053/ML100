載入套件

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets
np.random.seed(5)
%matplotlib inline
```

載入 toy example 資料集

```python
toy = datasets.make_blobs(centers=3, n_features=4,random_state=42)  #make_blobs聚類數據生成
X = toy[0]
y = toy[1]
```

設定 模型 估計參數

```python
estimators = [('k_means_8', KMeans(n_clusters=8)),
              ('k_means_3', KMeans(n_clusters=3)),
              ('k_means_bad_init', KMeans(n_clusters=3, n_init=1,init='random'))]
```

資料建模 並 視覺化 結果

```python
fignum = 1
titles = ['8 clusters', '3 clusters', '3 clusters, bad initialization']  #三種分群方式
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(4, 3)) #建立圖像的開頭
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=30, azim=134)  #elev=視角 azim=方位角視角
    
    ## fit data
    est.fit(X)
    labels = est.labels_ #取出標籤
    ax.scatter(X[:, 3], X[:, 0], X[:, 2],c=labels.astype(np.float),edgecolor='k')  #取X的第4,第1,第3個特徵
    ax.w_xaxis.set_ticklabels(['X'])
    ax.w_yaxis.set_ticklabels(['Y'])
    ax.w_zaxis.set_ticklabels(['Z'])
    ax.set_title(titles[fignum - 1]) #每張圖的title
    ax.dist = 10  #越大代表物體越遠,默認10
    fignum = fignum + 1
    
    
# Plot the ground truth
fig = plt.figure(fignum, figsize=(4, 3))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=30, azim=134)
for name, label in [('class0', 0),
                    ('class1', 1),
                    ('class2', 2)]:
    ax.text3D(X[y == label, 3].mean(),       #標記的記號(文字) X軸
              X[y == label, 0].mean(),       #Y軸
              X[y == label, 2].mean() +4,    #Z軸
              name,                          #要放什麼文字
              horizontalalignment='center',  #文字框框要偏哪(left/right/center)
              bbox=dict(alpha=.2, edgecolor='k', facecolor='k')) #文字框框 edge是邊緣 face是背景
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title('Ground Truth')
ax.dist = 10

#fig.show()
```
